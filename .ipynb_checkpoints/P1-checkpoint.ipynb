## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def extrapolate_lines(slopes,intercepts,ysize,top_proportion):
    """
    NOTE: this functions extrapolates the left and right lines it creates two line clusters: one 
    composed by lines with positive slope, and the other composed by lines with negative slope.
    
    Then it extracts the median of the slopes and intercepts in both clusters, and finally it returns the coordinates
    to plot the left and right lines, if there are enough values to get a good extrapolation of the lines, a boolean
    "lines_exist = True" is returned, but if there are not enough values in both sides, the boolean "lines_exist=False"
    which means that the lines can't be drawn
    
    %%INPUTS:  
    slopes: List with the slopes of all the lines found by hough
    intercepts: List with the intercepts of all the lines found by hough
    ysize: height of the image where the lines will be drawn
    
    %%OUTPUTS
    lines_exist: A boolean that is False if there are not enough lines to extra, so they can't be draw.
    line_coords: A list with 4 points, the x,y coordinates of the points to draw the lines in the order:
                    [(x1_line1,y1_line1),(x2_line1,y2_line1),(x1_line2,y1_line2),(x2_line2,y2_line2)]
    
    """
    lines_exist=False
    line_coords=list()
    
    if len(slopes)>4:
        index_positives=[cont for cont in range(len(slopes)) if  slopes[cont]>0]
        slopes_positive=np.take(slopes, index_positives)
        slopes_negative=np.delete(slopes,index_positives)
        if len(slopes_positive)>1 and len(slopes_negative)>1:
            lines_exist=True
            intercepts_positive=np.take(intercepts, index_positives)
            intercepts_negative=np.delete(intercepts,index_positives)
            median_positive=np.median(slopes_positive)
            median_negative=np.median(slopes_negative)
            median_positive_inter=np.median(intercepts_positive)
            median_negative_inter=np.median(intercepts_negative)
            x1_l1,y1_l1=((ysize*top_proportion-median_negative_inter)/median_negative,ysize*top_proportion)
            x2_l1,y2_l1=((ysize-median_negative_inter)/median_negative,ysize)
            x1_l2,y1_l2=((ysize*top_proportion-median_positive_inter)/median_positive,ysize*top_proportion)
            x2_l2,y2_l2=((ysize-median_positive_inter)/median_positive,ysize)
            line_coords=[(int(x1_l1),int(y1_l1)),(int(x2_l1),int(y2_l1)),(int(x1_l2),int(y1_l2)),(int(x2_l2),int(y2_l2))]
    return lines_exist,line_coords

def draw_lines(img, lines,top_proportion, color=[255, 0, 0], thickness=2,average_lines=True):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    ysize,xsize,_=img.shape
    slopes=list()
    intercepts=list()
    for line in lines:
        for x1,y1,x2,y2 in line:
            if x2-x1!=0:
                (slope,intercept) = np.polyfit((x1,x2), (y1,y2), 1)
                if slope>0.5 or  slope <-0.5:
                    slopes.append(slope)
                    intercepts.append(intercept)
                    if average_lines==False:
                        cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    if average_lines==True:
        line_exist,line_coords=extrapolate_lines(slopes,intercepts,ysize,top_proportion)
        if line_exist:
            cv2.line(img, line_coords[0], line_coords[1], color=(255,0,0), thickness=thickness)
            cv2.line(img, line_coords[2], line_coords[3], color=(255,0,0), thickness=thickness)
    return 

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap,top_proportion,average_lines=True):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines,top_proportion,thickness=5,average_lines=average_lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.
initial_images=['whiteCarLaneSwitch.jpg',
 'solidYellowLeft.jpg',
 'solidWhiteRight.jpg',
 'solidWhiteCurve.jpg',
 'solidYellowCurve2.jpg',
 'solidYellowCurve.jpg']
for imagename in initial_images:
    tini=time.time()
    image = mpimg.imread('test_images/'+ imagename)
    plt.figure()
    plt.title('original image')
    plt.imshow(image)
    duplicate_image = np.copy(image)
    #grayscale image
    gray_image=grayscale(duplicate_image)
    plt.figure()
    plt.title('grayscaled image')
    plt.imshow(gray_image,cmap='gray')
    cv2.imwrite("test_images/gray_"+imagename,gray_image)
    # gausian blur
    blured=gaussian_blur(gray_image, kernel_size=5)
    cv2.imwrite("test_images/blur_"+imagename,blured)
    # canny of image
    high_threshold=60
    low_threshold=20
    canny_image=canny(blured, low_threshold, high_threshold)
    cv2.imwrite("test_images/canny_"+imagename,canny_image)
    plt.figure()
    plt.title('canny image')
    plt.imshow(canny_image,cmap='gray')
    # roi image
    top_proportion=8/14
    (ysize,xsize)=canny_image.shape
    vertices = np.array([[[xsize/15,ysize],[xsize/2-xsize/80, ysize*top_proportion], [xsize/2+xsize/80, ysize*top_proportion], [xsize*14/15,ysize]]], dtype=np.int32)
    masked_image=region_of_interest(canny_image, vertices)
    plt.figure()
    plt.title('masked_image image')
    plt.imshow(masked_image,cmap='gray')
    cv2.imwrite("test_images/roi_"+imagename,masked_image)
    #hough
    rho=1
    theta=np.pi/180
    threshold=int(ysize/40)
    min_line_len=int(ysize/30)
    max_line_gap=int(ysize/100)
    hough_image=hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap,top_proportion,average_lines=False)
    image_opencv_rgb=cv2.cvtColor(hough_image, cv2.COLOR_RGB2BGR)
    cv2.imwrite("test_images/houghLines_"+imagename,image_opencv_rgb)
    plt.figure()
    plt.title('hough lines image')
    plt.imshow(hough_image)
    # weighted img with lines
    weighted= weighted_img(hough_image, image, α=0.8, β=1., γ=0.)
    image_opencv_rgb=cv2.cvtColor(weighted, cv2.COLOR_RGB2BGR)
    cv2.imwrite("test_images/houghLinesOverImage_"+imagename,image_opencv_rgb)
    plt.figure()
    plt.title('weighted image with multiple lines')
    plt.imshow(weighted)

    #
    hough_image2=hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap,top_proportion,average_lines=True)
    weighted2= weighted_img(hough_image2, image, α=0.8, β=1., γ=0.)
    image_opencv_rgb=cv2.cvtColor(weighted2, cv2.COLOR_RGB2BGR)
    cv2.imwrite("test_images/mergedLinesOverImage_"+imagename,image_opencv_rgb)
    plt.figure()
    plt.title('weighted image with extrapolated_lines')
    plt.imshow(weighted2)
    t_elap=time.time()-tini
    print('time',t_elap)

````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    duplicate_image = np.copy(image)

    #grayscale image
    gray_image=grayscale(duplicate_image)
    blured=gaussian_blur(gray_image, kernel_size=5)
    # canny of image
    high_threshold=60
    low_threshold=20
    canny_image=canny(blured, low_threshold, high_threshold)
    # roi image
    (ysize,xsize)=canny_image.shape
    # the top of the ROI will be in the "y" coordinate top_proportion*ysize, the bottom will be in the "y"= ysize
    top_proportion=8/14
    vertices = np.array([[[xsize/15,ysize],[xsize/2-xsize/80, ysize*top_proportion], [xsize/2+xsize/80, ysize*top_proportion], [xsize*14/15,ysize]]], dtype=np.int32)
    masked_image=region_of_interest(canny_image, vertices)

    #hough
    rho=1
    theta=np.pi/180
    threshold=int(ysize/40)
    min_line_len=int(ysize/30)
    max_line_gap=int(ysize/100)
    hough_image=hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap,top_proportion)

    # weighted img
    result= weighted_img(hough_image, image, α=0.8, β=1., γ=0.)
    ysize,xsize,_=result.shape
    #print('sizeresults',ysize,xsize)
    #t_elap=time.time()-tini
    #print('time',t_elap)
    return result
````
